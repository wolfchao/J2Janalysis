```python
J2JOD_industry = J2JOD[['periodicity', 'industry_orig']].groupby('industry_orig').count()
J2JOD_industry = J2JOD_industry.rename(columns={'periodicity' : 'count'})
J2JOD_industry = J2JOD_industry.reset_index()
J2JOD_industry
```

```python
label_industry = pd.read_csv("G:\Dataset\j2jod\label_industry.csv")
label_industry = label_industry.rename(columns={'industry' : 'industry_orig'})

J2JOD_industry = J2JOD_industry.merge(label_industry, on=["industry_orig"])
J2JOD_industry
```

```python
J2JOD_industry_dest = J2JOD[['periodicity', 'industry']].groupby('industry').count()
J2JOD_industry_dest = J2JOD_industry_dest.rename(columns={'periodicity' : 'count'})
J2JOD_industry_dest = J2JOD_industry_dest.reset_index()
label_industry = pd.read_csv("G:\Dataset\j2jod\label_industry.csv")

J2JOD_industry_dest = J2JOD_industry_dest.merge(label_industry, on=["industry"])
J2JOD_industry_dest
```

```python
J2JOD_industry_change = J2JOD[['periodicity', 'industry_orig', 'industry']].groupby(['industry_orig', 'industry']).count()
J2JOD_industry_change = J2JOD_industry_change.rename(columns={'periodicity' : 'count'})
J2JOD_industry_change = J2JOD_industry_change.reset_index()
J2JOD_industry_change
```

```python
J2JOD_industry_change[J2JOD_industry_change['industry_orig'] == J2JOD_industry_change['industry']]
```

```python
J2JOD_industry['dest_count'] = J2JOD_industry_dest['count']
J2JOD_industry['remain in original dest'] = J2JOD_industry_change['count']
J2JOD_industry['remain in original dest %'] = J2JOD_industry['remain in original dest'] / J2JOD_industry['count'] * 100
J2JOD_industry['new to the industry %'] = (J2JOD_industry['dest_count'] - J2JOD_industry['remain in original dest']) / J2JOD_industry['dest_count'] * 100
J2JOD_industry['increase'] = J2JOD_industry['dest_count'] - J2JOD_industry['count']
J2JOD_industry['increase %'] = J2JOD_industry['increase'] / J2JOD_industry['count'] * 100
J2JOD_industry
```

```python
plt.bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['remain in original dest %'][1:len(J2JOD_industry)])
plt.show()
```

```python
plt.bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['increase'][1:len(J2JOD_industry)])
plt.show()
```

```python

fig, ax = plt.subplots(3, 2, sharex='col')

# axes[0].set_title('square')
fig.set_figwidth(20)
fig.set_figheight(15)

ax[0][0].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['count'][1:len(J2JOD_industry)])
ax[0][0].set_title('J2J original indestory amount')
ax[0][1].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['remain in original dest %'][1:len(J2JOD_industry)])
ax[0][1].set_title('J2J remain in original indestory %')

ax[1][0].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['dest_count'][1:len(J2JOD_industry)])
ax[1][0].set_title('J2J destination indestory amount')

ax[1][1].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['new to the industry %'][1:len(J2JOD_industry)])
ax[1][1].set_title('J2J new to the industry %')

ax[2][0].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['increase'][1:len(J2JOD_industry)])
ax[2][0].set_title('J2J increase in the industry')

ax[2][1].bar(J2JOD_industry['industry_orig'][1:len(J2JOD_industry)], J2JOD_industry['increase %'][1:len(J2JOD_industry)])
ax[2][1].set_title('J2J increase in the industry %')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(25)
fig.set_figheight(25)
X_axis = np.arange(1, len(J2JOD_race_industry['industry_orig'].unique()))
plt.bar(X_axis - 0.15, J2JOD_race_industry[J2JOD_race_industry['race'] == "A1"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'White')
plt.bar(X_axis - 0.1, J2JOD_race_industry[J2JOD_race_industry['race'] == "A2"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'Black or African American')
plt.bar(X_axis - 0.05, J2JOD_race_industry[J2JOD_race_industry['race'] == "A3"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'American Indian or Alaska Native')
plt.bar(X_axis, J2JOD_race_industry[J2JOD_race_industry['race'] == "A4"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'Asian')
plt.bar(X_axis + 0.05, J2JOD_race_industry[J2JOD_race_industry['race'] == "A5"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'Native Hawaiian or Other Pacific Islander')
plt.bar(X_axis + 0.1, J2JOD_race_industry[J2JOD_race_industry['race'] == "A7"]['percentage'][1:len(J2JOD_race_industry['industry_orig'].unique())], 0.1, label = 'Two or More Race Groups')
X_ticks = J2JOD_race_industry['industry_orig'].unique()
X_ticks = X_ticks[1:len(J2JOD_race_industry['industry_orig'].unique())]
plt.xticks(X_axis, X_ticks)
plt.legend()
plt.title("percentage leaving rate by race in different industry")

plt.show()
```
